```python
import numpy as np
import matplotlib.pyplot as plt

def polinomio_lagrange(x, y, evaluar_en):
    """
    Calcula el polinomio de Lagrange para interpolar un conjunto de puntos.

    :param x: Lista de coordenadas x de los puntos.
    :param y: Lista de coordenadas y de los puntos.
    :param evaluar_en: Puntos en los que evaluar el polinomio interpolante.
    :return: Lista de valores y del polinomio interpolante en los puntos dados.
    """
    resultado = []

    for evaluar_en_punto in evaluar_en:
        interpolacion = 0

        for i in range(len(x)):
            termino = y[i]

            for j in range(len(x)):
                if i != j:
                    termino *= (evaluar_en_punto - x[j]) / (x[i] - x[j])

            interpolacion += termino

        resultado.append(interpolacion)

    return resultado

# Ejemplo de uso:
if __name__ == "__main__":
    # Datos de ejemplo
    x_datos = np.array([1, 2.2, 3.4, 4.8, 6, 7])
    y_datos = np.array([2, 2.8, 3, 3.2, 4, 5])

    # Punto adicional de interpolaci칩n
    x_nuevo_punto = 5.4
    y_nuevo_punto = polinomio_lagrange(x_datos, y_datos, [x_nuevo_punto])[0]

    # Puntos en los que evaluar el polinomio interpolante
    x_evaluar = np.linspace(1, 7, 100)

    # Calcular el polinomio interpolante de Lagrange
    y_interpolacion = polinomio_lagrange(x_datos, y_datos, x_evaluar)

    # Mostrar resultados
    plt.scatter(x_datos, y_datos, label='Datos originales')
    plt.scatter(x_nuevo_punto, y_nuevo_punto, color='green', label='punto de interpolaci칩n')
    plt.plot(x_evaluar, y_interpolacion, color='red', label='Polinomio de Lagrange')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.title('Polinomio de Lagrange para Interpolaci칩n con Nuevo Punto')
    plt.show()
```
