```python
import numpy as np

def regla_de_cramer(matriz_coeficientes, vector_terminos_independientes):
    det_A = np.linalg.det(matriz_coeficientes)
    n = len(matriz_coeficientes)

    soluciones = []

    for i in range(n):
        matriz_sustituida = matriz_coeficientes.copy()
        matriz_sustituida[:, i] = vector_terminos_independientes
        det_Ai = np.linalg.det(matriz_sustituida)
        solucion_i = det_Ai / det_A
        soluciones.append(solucion_i)

    return soluciones

# Sistema de ecuaciones
A = np.array([
    [2, -3, 2, -4],
    [-1, -2, 3, 2],
    [2, 1, -4, -2],
    [-3, 4, -1, 1]
])

b = np.array([2.3, -3.0, 7.7, -7.0])

# Resolver utilizando la regla de Cramer
soluciones = regla_de_cramer(A, b)

# Imprimir resultados
print("Soluciones del sistema:")
for i, solucion in enumerate(soluciones):
    print(f"x{i+1} = {solucion}")
```
