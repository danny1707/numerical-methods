```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

def taylor_series(func, x0, n):
    """
    Aproximación de la función mediante la serie de Taylor.

    :param func: La función a aproximar.
    :param x0: Punto alrededor del cual se realiza la aproximación.
    :param n: Orden de la serie de Taylor (número de términos a considerar).
    :return: Aproximación de la función mediante la serie de Taylor.
    """
    x = sp.symbols('x')
    taylor_approx = func.series(x, x0, n).removeO()

    return taylor_approx

# Ejemplo de uso:
if __name__ == "__main__":
    # Definir la función y el punto de aproximación
    x = sp.symbols('x')
    func = sp.sin(x)
    point_of_approximation = 0
    order_of_series = 4

    # Calcular la aproximación de la serie de Taylor
    taylor_approximation = taylor_series(func, point_of_approximation, order_of_series)

    # Imprimir la aproximación de la serie de Taylor
    print(f"Aproximación de la serie de Taylor: {taylor_approximation}")

    # Crear gráfico para visualizar la aproximación
    x_values = np.linspace(-2*np.pi, 2*np.pi, 100)
    func_values = np.sin(x_values)
    taylor_values = [taylor_approximation.subs('x', val) for val in x_values]

    plt.plot(x_values, func_values, label="sin(x)")
    plt.plot(x_values, taylor_values, label=f"Serie de Taylor (orden {order_of_series})")
    plt.title("Aproximación de la Serie de Taylor para sin(x)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```
