```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -x**3 - np.cos(x)

def secant_method(p0, p1, tol, max_iter):
    iter_count = 0
    while iter_count < max_iter:
        f_p0 = f(p0)
        f_p1 = f(p1)

        p2 = p1 - f_p1 * (p1 - p0) / (f_p1 - f_p0)

        if abs(p2 - p1) < tol:
            return p2, iter_count + 1

        p0, p1 = p1, p2
        iter_count += 1

    raise Exception("El método de la secante no convergió después de {} iteraciones".format(max_iter))

# Parámetros iniciales
p0 = -1
p1 = 0
tolerancia = 1e-8
max_iteraciones = 1000

# Llamada al método de la secante
raiz, iteraciones = secant_method(p0, p1, tolerancia, max_iteraciones)

# Imprimir resultado
print("La raíz aproximada es:", raiz)
print("Número de iteraciones:", iteraciones)

# Graficar la función
x_values = np.linspace(-2, 2, 400)
y_values = f(x_values)

plt.plot(x_values, y_values, label='$f(x)=-x^3 - \cos(x)$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--', label='Eje x')
plt.scatter([raiz], [0], color='red', label='Raíz encontrada')

# Configuración de la gráfica
plt.title('Gráfica de $f(x)=-x^3 - \cos(x)$ y la raíz encontrada')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
