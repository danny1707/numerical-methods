```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definir funciones para ajustar
def funcion_potencia(x, a, b):
    return a * x**b

def funcion_exponencial(x, a, b):
    return a * np.exp(b * x)

def funcion_cuadratica(x, a, b, c):
    return a * x**2 + b * x + c

# FunciÃ³n general para realizar el ajuste y graficar resultados
def ajuste_y_grafico(x, y, funcion, nombre_funcion):
    # Realizar el ajuste utilizando curve_fit
    parametros_optimos, covarianza = curve_fit(funcion, x, y)

    # Graficar datos y ajuste
    plt.scatter(x, y, label='Datos')
    plt.plot(x, funcion(x, *parametros_optimos), color='red', label=f'Ajuste {nombre_funcion}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.title(f'Ajuste de {nombre_funcion}')
    plt.show()

    # Mostrar los parÃ¡metros Ã³ptimos del ajuste
    print(f"ParÃ¡metros Ã³ptimos para el ajuste {nombre_funcion}: {parametros_optimos}")

# Ejemplo de uso:
if __name__ == "__main__":
    # Generar datos de ejemplo
    np.random.seed(0)
    x = np.linspace(0, 10, 100)
    y_potencia = 2 * x**1.5 + np.random.normal(scale=5, size=len(x))
    y_exponencial = 3 * np.exp(0.5 * x) + np.random.normal(scale=5, size=len(x))
    y_cuadratica = 1.5 * x**2 + 2 * x + 1 + np.random.normal(scale=5, size=len(x))

    # Ajuste y grÃ¡fico para funciÃ³n de potencia
    ajuste_y_grafico(x, y_potencia, funcion_potencia, 'Potencia')

    # Ajuste y grÃ¡fico para funciÃ³n exponencial
    ajuste_y_grafico(x, y_exponencial, funcion_exponencial, 'Exponencial')

    # Ajuste y grÃ¡fico para funciÃ³n cuadrÃ¡tica
    ajuste_y_grafico(x, y_cuadratica, funcion_cuadratica, 'CuadrÃ¡tica')
```
