```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

def herramientas_estadisticas(datos, confianza=0.95):
    """
    Calcula varias herramientas estadísticas para un conjunto de datos, incluyendo el intervalo de confianza.

    :param datos: Lista o array de datos numéricos.
    :param confianza: Nivel de confianza para el intervalo de confianza (por defecto, 0.95).
    :return: Media, desviación estándar, varianza, coeficiente de variación, variación y intervalo de confianza.
    """
    datos = np.array(datos)

    # Media aritmética
    media = np.mean(datos)

    # Desviación estándar
    desviacion_estandar = np.std(datos)

    # Varianza
    varianza = np.var(datos)

    # Coeficiente de variación
    coeficiente_variacion = (desviacion_estandar / media) * 100  # Porcentaje

    # Variación
    variacion = desviacion_estandar**2

    # Intervalo de confianza para la media
    n = len(datos)
    t_valor = t.ppf((1 + confianza) / 2, df=n - 1)
    error_estandar = desviacion_estandar / np.sqrt(n)
    intervalo_confianza = t_valor * error_estandar

    return media, desviacion_estandar, varianza, coeficiente_variacion, variacion, intervalo_confianza

# Ejemplo de uso:
if __name__ == "__main__":
    # Datos de ejemplo
    datos = [2, 4, 4, 4, 5, 5, 7, 9]

    # Calcular herramientas estadísticas
    media, desviacion_estandar, varianza, coeficiente_variacion, variacion, intervalo_confianza = herramientas_estadisticas(datos)

    # Mostrar resultados
    print(f"Media aritmética: {media}")
    print(f"Desviación estándar: {desviacion_estandar}")
    print(f"Varianza: {varianza}")
    print(f"Coeficiente de variación: {coeficiente_variacion}%")
    print(f"Variación: {variacion}")
    print(f"Intervalo de confianza para la media: [{media - intervalo_confianza}, {media + intervalo_confianza}]")

    # Gráficos
    plt.figure(figsize=(12, 8))

    # Histograma de los datos
    plt.subplot(2, 2, 1)
    plt.hist(datos, bins=range(min(datos), max(datos) + 1), edgecolor='black')
    plt.title('Histograma de Datos')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')

    # Boxplot
    plt.subplot(2, 2, 2)
    plt.boxplot(datos, vert=False)
    plt.title('Boxplot de Datos')

    # Gráfico de barras para la media y desviación estándar
    plt.subplot(2, 2, 3)
    plt.bar(['Media', 'Desv. Estándar'], [media, desviacion_estandar], color=['blue', 'orange'])
    plt.title('Media y Desviación Estándar')

    # Gráfico de barras para el coeficiente de variación y variación
    plt.subplot(2, 2, 4)
    plt.bar(['Coef. Variación', 'Variación'], [coeficiente_variacion, variacion], color=['green', 'red'])
    plt.title('Coeficiente de Variación y Variación')

    plt.tight_layout()
    plt.show()
```
