```python
def synthetic_division(coefficients, c):
    """
    Realiza la división sintética de un polinomio por (x - c).

    :param coefficients: Lista de coeficientes del polinomio en orden descendente.
    :param c: Valor de 'c' en (x - c).
    :return: Coeficientes del cociente y el residuo.
    """
    n = len(coefficients) - 1
    result = list(coefficients)

    for i in range(n - 1, 0, -1):
        result[i - 1] += c * result[i]

    remainder = result[0]
    result[0] += c * remainder

    return result[1:], remainder

# Ejemplo de uso:
if __name__ == "__main__":
    # Coeficientes del polinomio (3x^3 - 4x^2 + 2x - 5)
    coefficients = [3, -4, 2, -5]

    # Valor de 'c' en (x - c)
    c_value = 2

    # Realizar la división sintética
    quotient, remainder = synthetic_division(coefficients, c_value)

    # Mostrar resultados
    print(f"Cociente: {quotient}")
    print(f"Residuo: {remainder}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def evaluate_polynomial(coefficients, x):
    """Evalúa el polinomio con los coeficientes dados en el punto x."""
    return np.polyval(coefficients, x)

# Coeficientes del polinomio (3x^3 - 4x^2 + 2x - 5)
original_coefficients = [3, -4, 2, -5]

# Valor de 'c' en (x - c)
c_value = 2

# Realizar la división sintética
quotient, remainder = synthetic_division(original_coefficients, c_value)

# Crear datos para graficar el polinomio original y el resultado de la división
x_values = np.linspace(-5, 5, 400)
y_original = evaluate_polynomial(original_coefficients, x_values)
y_quotient = evaluate_polynomial(quotient, x_values - c_value)

# Graficar el polinomio original y el resultado de la división sintética
plt.plot(x_values, y_original, label="Polinomio Original")
plt.plot(x_values, y_quotient, label="Cociente (división sintética)")
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.title("División Sintética y Polinomio Original")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
