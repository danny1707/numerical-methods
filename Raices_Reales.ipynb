```python
import numpy as np
import matplotlib.pyplot as plt

def funcion_a(x):
    return -26 + 85*x - 91*x**2 + 44*x**3 - 8*x**4 + x**5

def funcion_b(x):
    return (0.8 - 0.3*x) / x

# Rango de valores para x
x_vals = np.linspace(-10, 10, 400)

# Calcular los valores de y para cada función
y_vals_a = funcion_a(x_vals)
y_vals_b = funcion_b(x_vals)

# Encontrar raíces
raices_a = np.roots([-26, 85, -91, 44, -8, 1])
raices_b = np.roots([-0.3, 0.8])

# Crear gráficos para cada función
plt.figure(figsize=(12, 6))

# Función a
plt.subplot(1, 2, 1)
plt.plot(x_vals, y_vals_a, label='f(x) = -26 + 85x - 91x^2 + 44x^3 - 8x^4 + x^5')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--', label='y=0')
plt.scatter(raices_a, [0]*len(raices_a), color='red', label='Raíces')
plt.title('Función a')
plt.legend()
plt.grid(True)

# Función b
plt.subplot(1, 2, 2)
plt.plot(x_vals, y_vals_b, label='f(x) = (0.8 - 0.3x)/x')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--', label='y=0')
plt.scatter(raices_b, [0]*len(raices_b), color='red', label='Raíces')
plt.title('Función b')
plt.legend()
plt.grid(True)

# Imprimir raíces
print("Raíces de la función a:", raices_a)
print("Raíces de la función b:", raices_b)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
