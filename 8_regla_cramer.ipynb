```python
import numpy as np

def cramer_rule(A, B):
    """
    Resuelve un sistema de ecuaciones lineales Ax = B utilizando la regla de Cramer.

    :param A: Matriz de coeficientes del sistema.
    :param B: Vector de términos constantes.
    :return: Vector x que satisface Ax = B.
    """
    det_A = np.linalg.det(A)
    n = len(B)
    x = np.zeros(n)

    for i in range(n):
        Ai = A.copy()
        Ai[:, i] = B
        det_Ai = np.linalg.det(Ai)
        x[i] = det_Ai / det_A

    return x

# Ejemplo de uso:
if __name__ == "__main__":
    # Definir el sistema de ecuaciones lineales: 2x + y - z = 8, -3x - y + 2z = -11, -2x + y + 2z = -3
    A = np.array([[2, 1, -1],
                  [-3, -1, 2],
                  [-2, 1, 2]])

    B = np.array([8, -11, -3])

    # Resolver el sistema de ecuaciones lineales utilizando la regla de Cramer
    solution = cramer_rule(A, B)

    # Mostrar la solución
    print("Solución del sistema de ecuaciones lineales:")
    for i, val in enumerate(solution):
        print(f"x_{i+1} = {val}")
```
