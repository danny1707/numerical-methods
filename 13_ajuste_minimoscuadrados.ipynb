```python
import numpy as np
import matplotlib.pyplot as plt

def ajuste_minimos_cuadrados(x, y, grado):
    """
    Realiza el ajuste de funciones por mÃ­nimos cuadrados para un conjunto de datos.

    :param x: Valores x de los datos.
    :param y: Valores y de los datos.
    :param grado: Grado del polinomio de ajuste.
    :return: Coeficientes del polinomio de ajuste.
    """
    # Crear matriz de diseÃ±o
    X = np.vander(x, grado + 1)

    # Calcular coeficientes por mÃ­nimos cuadrados
    coeficientes = np.linalg.lstsq(X, y, rcond=None)[0]

    return coeficientes

def evaluar_polinomio(coeficientes, x):
    """
    EvalÃºa un polinomio dados sus coeficientes y un valor x.

    :param coeficientes: Coeficientes del polinomio.
    :param x: Valor x.
    :return: Valor y del polinomio en x.
    """
    return np.polyval(coeficientes, x)

# Ejemplo de uso:
if __name__ == "__main__":
    # Generar datos de ejemplo
    np.random.seed(0)
    x = np.linspace(0, 10, 20)
    y = 2 * x + 1 + np.random.normal(scale=2, size=len(x))

    # Ajuste de funciones por mÃ­nimos cuadrados con un polinomio de grado 1 (lineal)
    grado_del_polinomio = 1
    coeficientes = ajuste_minimos_cuadrados(x, y, grado_del_polinomio)

    # Evaluar el polinomio en nuevos puntos
    x_evaluacion = np.linspace(0, 10, 100)
    y_prediccion = evaluar_polinomio(coeficientes, x_evaluacion)

    # Mostrar resultados
    plt.scatter(x, y, label='Datos')
    plt.plot(x_evaluacion, y_prediccion, color='red', label=f'Ajuste polinÃ³mico (grado {grado_del_polinomio})')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.title('Ajuste de funciones por mÃ­nimos cuadrados')
    plt.show()

    print(f"Coeficientes del polinomio de ajuste: {coeficientes}")
```
